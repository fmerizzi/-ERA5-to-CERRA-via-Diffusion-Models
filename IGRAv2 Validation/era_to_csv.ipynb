```python
import xarray as xr
import cfgrib
import numpy as np 
import netCDF4
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
```

```python
era = "/home/stefano/Public/climate_comparison/data/era5_windspeed_2009.grib"
ds_grib_era = xr.open_dataset(era)
```

```python
ds_grib_era
```

```python
df_era = ds_grib_era.to_dataframe()
```

```python
df_era.head()
```

```python
df_era = df_era.reset_index()
df_era['wind_speed'] = np.sqrt(df_era.u10**2 + df_era.v10**2)
df_era = df_era.drop(columns=['step', 'number','surface', "valid_time", "u10", "v10"])
```

```python
df_era_first = df_era.iloc[:2704]
# keep only the values of the wind speed
df_era_first = df_era_first.drop(columns=['latitude', 'longitude'])
# transform the dataframe wind_speed column into an array
first_array = df_era_first.to_numpy()
first_arraylist = []
for i in range(0,2704):
    first_arraylist.append(first_array[i][1])
first_array = np.array(first_arraylist)
first_array = first_array.reshape(52,52)
plt.imshow(first_array)
plt.imsave('test.png', first_array)
```
